```python
import torch

print("Starting pipeline test...")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

if device.type == "cuda":
    print("GPU:", torch.cuda.get_device_name(0))
    print("VRAM:", torch.cuda.get_device_properties(0).total_memory/1e9, "GB")

print("\nBuilding model...")

model = TextToVideoPipeline(
    base_ch=64,
    ch_mult=(1,2,2),
    max_frames=4,
    num_diff_steps=100
).to(device)

params = sum(p.numel() for p in model.parameters())
print("Parameters:", params/1e6, "Million")

print("\nCreating dummy inputs...")

video = torch.randn(1,3,4,64,64).to(device)
tokens = torch.randint(0,49408,(1,10)).to(device)

print("Running training loss...")

loss = model.training_loss(video, tokens)

print("Loss:", loss.item())

print("\nGenerating video...")

model.eval()

out = model.generate(
    tokens,
    num_frames=4,
    height=8,
    width=8,
    num_steps=5,
    seed=42
)

print("Output shape:", out.shape)

print("\nPipeline working correctly ✅")
```

```python
import torchvision
import imageio

def save_video(video_tensor, filename="output.mp4", fps=4):

    video = video_tensor[0].cpu()
    video = (video + 1) / 2
    video = video.clamp(0,1)

    frames = []

    for i in range(video.shape[1]):
        frame = video[:,i]
        frame = frame.permute(1,2,0).numpy()
        frame = (frame*255).astype("uint8")
        frames.append(frame)

    imageio.mimsave(filename, frames, fps=fps)
```
