```python
import torchvision
import imageio

def save_video(video_tensor, filename="output.mp4", fps=4):

    video = video_tensor[0].cpu()
    video = (video + 1) / 2
    video = video.clamp(0,1)

    frames = []

    for i in range(video.shape[1]):
        frame = video[:,i]
        frame = frame.permute(1,2,0).numpy()
        frame = (frame*255).astype("uint8")
        frames.append(frame)

    imageio.mimsave(filename, frames, fps=fps)
```

```python
import cv2
import torch

def extract_frames(video_path, num_frames=4):

    cap = cv2.VideoCapture(video_path)

    frames = []

    while len(frames) < num_frames:

        ret, frame = cap.read()

        if not ret:
            break

        frame = cv2.resize(frame,(64,64))
        frame = frame[:,:,::-1]   # BGR → RGB

        frames.append(frame)

    cap.release()

    frames = torch.tensor(frames).permute(3,0,1,2).float()

    frames = frames/127.5 - 1

    return frames
```

```python
from torch.utils.data import Dataset

class VideoDataset(Dataset):

    def __init__(self, caption_file):

        self.data = []

        with open(caption_file,"r") as f:

            for line in f:

                video,caption = line.strip().split("|")

                self.data.append((video,caption))

    def __len__(self):
        return len(self.data)

    def __getitem__(self,idx):

        video_name,caption = self.data[idx]

        video_path = f"/content/{video_name}"

        frames = extract_frames(video_path)

        tokens = torch.randint(0,49408,(10,))

        return frames, tokens
```

```python
import cv2
import torch

def extract_frames(video_path, num_frames=4):

    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        raise ValueError(f"Cannot open video: {video_path}")

    frames = []

    while True:

        ret, frame = cap.read()

        if not ret:
            break

        frame = cv2.resize(frame,(64,64))
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        frames.append(frame)

        if len(frames) >= num_frames:
            break

    cap.release()

    if len(frames) == 0:
        raise ValueError("No frames extracted from video")

    frames = torch.tensor(frames).float()
    frames = frames.permute(3,0,1,2)

    frames = frames/127.5 - 1

    return frames
```
